AQI Distribution Analysis

The AQI values exhibit a positively skewed distribution. Most observations fall between 40 and 150 AQI, while a small number of extreme observations produce a long right tail. Boxplot analysis also indicates the presence of outliers, which likely correspond to severe pollution events. These observations were retained during model training because they represent rare but important real-world conditions.

```python
corr=df.corr(numeric_only=True)["US_AQI"].sort_values(ascending=False)
print(corr)
```

```python
city_aqi=df.groupby("City")["US_AQI"].mean().sort_values(ascending=True)
city_aqi
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

city_aqi.plot(kind="bar")

plt.title("Average AQI by City")
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.xticks(rotation=45)

plt.show()
```

```python
month_aqi = (
    df.groupby("Month")["US_AQI"]
      .mean()
)

print(month_aqi)
```

```python
plt.figure(figsize=(10,5))

month_aqi.plot(
    marker='o',
    linewidth=2
)

plt.title("Average AQI by Month")
plt.xlabel("Month")
plt.ylabel("Average AQI")

plt.grid(True)

plt.show()
```
